```python
#under the assumption we are replacing the bathroom null value with the condition if bedroom >1 replace with bedroomvalue - 1 , else 1  => decided based on date build also

def fill_null_bath(i):
    if pd.isnull(i["N_BATHROOM"]):
        if i["N_BEDROOM"]>1:
            return i["N_BEDROOM"]-1
        else:
            return 1
    else:
        return i["N_BATHROOM"]
```

```python
df["N_BATHROOM"]=df.apply(lambda x:fill_null_bath(x),axis=1)
df
```

```python
#2 # next col checking the null values
# [check null] we are replaceing the null values of col QS_OVERALL with the average of the QS_ROOMS QS_BATHROOM QS_BEDROOM colums
avg_cols =["QS_ROOMS","QS_BATHROOM","QS_BEDROOM"]
tar_col = "QS_OVERALL"
def fill_null(i):
    if pd.isnull(i[tar_col]):
        return i[avg_cols].sum()/len(avg_cols)    # return i[avg_cols].mean()
    else:
        return i[tar_col]
```

```python
df[tar_col]=df.apply(lambda x:fill_null(x),axis=1)
df
```

```python
#5 if data / time occurs - convert into pandas understandable format - further data transformation is advisable.

# Converting Date/time format because DATE_SALE & DATE_BUILD was object data type so i am changing to Date/time format.
df["DATE_SALE"] = pd.to_datetime(df["DATE_SALE"],format="%d-%m-%Y")
df["DATE_BUILD"] = pd.to_datetime(df["DATE_BUILD"],format="%d-%m-%Y")
```

```python
# to see difference b/w date build year to current year so i am adding "build_age".

from datetime import datetime
def build_age(timestamp):
    today = datetime.today()
    temp = timestamp
    return today.year - temp.year
df["DATE_BUILD"] = df["DATE_BUILD"].apply(build_age)

df
```

```python
# Date_sale task done
# if this data type was in object we can't do if is in pandas also.
# df1["DATE_SALE"].dt.year                    # this is splitting year query
df["YEAR_SALE"]=df["DATE_SALE"].dt.year     # column associated
df["MONTH_SALE"]=df["DATE_SALE"].dt.month
df["DAY_SALE"]=df["DATE_SALE"].dt.day
df["WEEK_SALE"]=df["DATE_SALE"].dt.day_of_week  # .dt.week will tell which week 0 to 6 days are there will tell anyone
# df1["QUARTER_SALE"]=df1["DATE_SALE"].dt.quarter
df["WEEK_SALE"]=df["DATE_SALE"].dt.weekday

df
```

```python
# After this Need to split x & y variables [Independent & dependent] then standardization want to do.

x=df.drop("SALES_PRICE", axis=1)
y=df["SALES_PRICE"]
x
```

```python
y
```

```python
# Before standardization want to corrla. I did it above.

# standardization
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# In x_train thought process i am doing standardization [code change in this line "scaler.fit(x_train)"] and we can do bethis step also while spliting  x = Independent & y = dependent variables columns.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
scaler=StandardScaler()
scaler.fit(x_train)
x_train_scaler = scaler.transform(x_train)
x_test_scaler = scaler.transform(x_test)
```

```python
# pip install tensorflow
import tensorflow as tf

# dir(tf.keras.metrics)

# I/p layer 22 --> HL1(810) --> HL2(512) --> HL3(256) --> HL4(128) --> HL5(64) --> HL6(16) --> O/p()
# Activation = LeakyReLU
```

```python
x_train_scaler.shape[1]

# x_train_scaler.shape[1] Directly we can enter in code because we are hardcoding values it may be enter wrongly.
```

```python
# understand how many metrics are there.
# R2Score, MeanAbsoluteError.
```

```python
# for continues data O/P activation need to remove.
# loss="mean_squared_error" need to add it is advisable.

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model=Sequential()
model.add(Dense(724,input_dim=x_train_scaler.shape[1], activation=tf.keras.layers.LeakyReLU(alpha=0.01)))
model.add(Dense(512,input_dim=x_train_scaler.shape[1], activation=tf.keras.layers.LeakyReLU(alpha=0.01)))
model.add(Dense(256,input_dim=x_train_scaler.shape[1], activation=tf.keras.layers.LeakyReLU(alpha=0.01)))
model.add(Dense(64,input_dim=x_train_scaler.shape[1], activation=tf.keras.layers.LeakyReLU(alpha=0.01)))
model.add(Dense(8,input_dim=x_train_scaler.shape[1], activation=tf.keras.layers.LeakyReLU(alpha=0.01)))
model.add(Dense(1))
model.compile(loss="mean_squared_error", optimizer="adam", metrics=[tf.keras.metrics.R2Score(), tf.keras.metrics.MeanAbsoluteError(), tf.keras.metrics.MeanAbsolutePercentageError(), tf.keras.metrics.MeanSquaredError()])
```

```python
result=model.fit(x_train_scaler, y_train, epochs=25, validation_split=0.12, batch_size=16)
```

```python
result_history_df = pd.DataFrame(result.history)
result_history_df
```
